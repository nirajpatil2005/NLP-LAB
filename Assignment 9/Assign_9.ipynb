Load model

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

model_name = "nlptown/bert-base-multilingual-uncased-sentiment"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

**Indian language input**

```python
texts = [
    "यह फिल्म बहुत अच्छी है",      # Hindi - Positive
    "सेवा बहुत खराब थी",           # Hindi - Negative
    "हा मोबाइल खूप छान आहे",        # Marathi - Positive
    "हा अनुभव फारच वाईट होता"      # Marathi - Negative
]
```

**Predict sentiment**

```python
inputs = tokenizer(texts, padding=True, truncation=True, return_tensors="pt")

with torch.no_grad():
    outputs = model(**inputs)

scores = torch.softmax(outputs.logits, dim=1)
predictions = torch.argmax(scores, dim=1)
```

**Map labels**

```python
for text, pred in zip(texts, predictions):
    stars = pred.item() + 1
    if stars <= 2:
        sentiment = "Negative"
    elif stars == 3:
        sentiment = "Neutral"
    else:
        sentiment = "Positive"

    print(text)
    print("Sentiment:", sentiment, f"({stars} stars)\n")
```
