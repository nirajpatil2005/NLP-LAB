Assignment no 8. Apply WordNet-based Word Sense Disambiguation to improve meaning interpretation in
ambiguous sentences.

**Install & download WordNet data**

```python
import nltk

nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt')
```

**Import required modules**

```python
from nltk.corpus import wordnet as wn
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize
```

**Basic Word Sense Disambiguation (Lesk)**

```python
nltk.download('punkt_tab')
sentence = "I went to the bank to deposit money"
tokens = word_tokenize(sentence)

sense = lesk(tokens, "bank")

print(sense)
print(sense.definition())
```

**Compare with another meaning**

```python
sentence2 = "The fisherman sat on the bank of the river"
tokens2 = word_tokenize(sentence2)

sense2 = lesk(tokens2, "bank")

print(sense2)
print(sense2.definition())
```

**Make it reusable (function)**

```python
def wordnet_wsd(sentence, target_word):
    tokens = word_tokenize(sentence)
    sense = lesk(tokens, target_word)

    if sense:
        return {
            "word": target_word,
            "synset": sense.name(),
            "definition": sense.definition(),
            "examples": sense.examples()
        }
    else:
        return "No sense found"
```

**Test on multiple ambiguous words**

```python
sentences = [
    ("He cracked the bat and hit a six", "bat"),
    ("The bat flew out of the cave", "bat"),
    ("She wore a beautiful ring", "ring"),
    ("The phone began to ring loudly", "ring")
]

for s, w in sentences:
    print("\nSentence:", s)
    print(wordnet_wsd(s, w))
```
